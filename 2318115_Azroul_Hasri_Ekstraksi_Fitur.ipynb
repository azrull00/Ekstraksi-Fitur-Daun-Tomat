# **Tugas Ekstraksi Fitur Daun Tomat**

```python
#saya disini langsung import dataset dari kaggle menggunakan library kagglehub

import kagglehub


path = kagglehub.dataset_download("kaustubhb999/tomatoleaf")

print("Path to dataset files:", path)
```

```python
#import library lain yang dibutuhkan

import os
import cv2
import numpy as np
import pandas as pd
```

```python
# 1. Mencari 10 citra pertama dari direktori dataset yang sudah didownload
image_paths = []
for root, dirs, files in os.walk(path): #ambil dari path dataset yg sudah saya definisikan diatas
    for file in files:
        # Memastikan file adalah gambar
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_paths.append(os.path.join(root, file))
            # menghentikan pencarian jika sudah dapat 10 citra
            if len(image_paths) == 10:
                break
    if len(image_paths) == 10:
        break

data_fitur = []

print(f"Ditemukan {len(image_paths)} citra. Memulai ekstraksi fitur \n")
```

```python
# 2 & 3. Loop untuk memproses 10 citra dan ekstraksi HSV
for img_path in image_paths:
    file_name = os.path.basename(img_path)

    # Baca citra menggunakan OpenCV
    img_bgr = cv2.imread(img_path)

    if img_bgr is not None:
        # Menyamakan ukuran menjadi 224x224 pixel agar seragam (Syarat Poin 1)
        img_resized = cv2.resize(img_bgr, (224, 224))

        # Konversi warna dari BGR (default OpenCV) ke HSV
        img_hsv = cv2.cvtColor(img_resized, cv2.COLOR_BGR2HSV)

        # Ekstraksi Fitur: Menghitung nilai rata-rata (Mean) untuk channel H, S, dan V
        mean_h = np.mean(img_hsv[:, :, 0])
        mean_s = np.mean(img_hsv[:, :, 1])
        mean_v = np.mean(img_hsv[:, :, 2])

        # Simpan ke dalam list
        data_fitur.append({
            'Nama File': file_name,
            'Mean Hue': mean_h,
            'Mean Saturation': mean_s,
            'Mean Value': mean_v
        })
        print(f"✅ Berhasil diekstrak: {file_name}")

# Membuat DataFrame dari data_fitur menggunakan Pandas
df = pd.DataFrame(data_fitur)

# Tampilkan tabel hasil di Colab
print("\n=== Tabel Hasil Ekstraksi Fitur HSV ===")
display(df)
```

```python
# 4. Menyimpan hasil ekstraksi ke file .csv dan .xlsx

file_csv = '/content/Ekstraksi_Fitur_Daun_Tomat.csv'
file_xlsx = '/content/Ekstraksi_Fitur_Daun_Tomat.xlsx'

df.to_csv(file_csv, index=False)
df.to_excel(file_xlsx, index=False)

print(f"\n🎉 Proses Selesai! File berhasil disimpan di:")
print(f"👉 {file_csv}")
print(f"👉 {file_xlsx}")
```
